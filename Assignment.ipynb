#### Starting off, we need to base our analysis based on the requirment or the sought outcome, which is the target we are focusing on, the price. We are doing our analysis based on what factors affect the price of our property.

```python
df_1[["price","sqft_living","condition"]].groupby("condition").mean()
```

```python
df_1[["price","grade"]].groupby("grade").mean().round()
```

```python
df_1[["price","bathrooms"]].groupby("bathrooms").mean().round()
```

```python
df_2 = df_1.copy()
df_2 = df_2[["price", "sqft_living"]]
df_2["sqft_living_range"] = pd.cut(df_2["sqft_living"],bins=10)
df_2[["price","sqft_living_range"]].groupby("sqft_living_range").mean().round().reset_index()
```

#### Here we used the .cut() method in order to produce ranges for the sqft_living for the properties. This showed the incremetnal increase of the price as the ranges increases.

view, waterfront, agg, how to make agg tables more clear,range the lat and long corr

```python
df[["view","waterfront","price"]].groupby(["waterfront","view"]).mean().round()
```

#### We can see in this dataset that view and waterfront are impacting the price. Firstly we have no waterfronts with view starting from zero till 4 and secondly having a waterfront with views from 1 to 4. The prices are considerably higher when we have a water front and view with a higher value

```python
df_2 = df.copy()
df_2 = df_2[["price", "lat", "long"]]
df_2["lat_range"] = pd.cut(df_2["lat"], bins=6)
df_2["long_range"] = pd.cut(df_2["long"], bins=6)
df_final=df_2[["price","lat_range","long_range"]].groupby(["lat_range","long_range"]).mean().round()
df_final.dropna().sort_values("price")
```

#### Organizing the display of agg method:

```python
def highlight_max(s):
    is_max = s == s.max()
    return ['background-color: yellow' if v else '' for v in is_max]
result = df[["price", "condition", "sqft_living"]].groupby("condition").agg(["mean", "min", "max"]).round()
styled_result = result.style.apply(highlight_max, axis=None)
styled_result
```

#### We can conclude from this table that depending on the location the price varies as well. considering the other factors as well would be a good idea to check if the location as well as the size of the property has an impact.

## Part 2, Correlation, Pearson Correlation Between Variables:

#### It measures the linear relationship between two variables. It is very important to note, that correlation doesn't mean causation.

#### If the correlation value is greater than zero, means positively correlated, means an increase in a certain variable will lead in the increase of the other.

#### Here is an example of a positive correlation between the price and sqft_living.

```python
df[["price","sqft_living"]].corr()
```

#### Example of no correlation: Here we can see that correlation between the price and the yr_renovated is close to zero.

```python
df[["price","yr_renovated"]].corr()
```

```python
df[["price","yr_built"]].corr()
```

The correlation is close to zero meaning there's no correlation between them.
